#### **<span style="color:blue">[2-1] air_21, air_22 의 '측정일시'를 활용하여 'time' 변수 생성</span>**

* air_21, air_22 의 '측정일시'를 활용하여 'time'변수 생성
* time 변수를 to_datetime으로 데이터 타입 변경
* 참고: 미세먼지 데이터는 1시-24시, 날씨 데이터는 0시-23시로 구성되어 있습니다. [2-5]에서 미세먼지와 날씨 데이터를 time 기준으로 합치려면 날씨 기준이 동일해야 합니다. 미세먼지 데이터에서 time 변수 생성 시 이를 미리 고려(예:측정일시 값 -1)하세요.

```python
from datetime import datetime, timedelta

def zerofrom24(datetime, _format):
    datetime = str(datetime)
    try:
        return pd.to_datetime(datetime, format = _format)
    except:
        datetime = datetime[:-2] + '00'
        return pd.to_datetime(datetime, format = _format)+ timedelta(days=1)
air_21['time']=air_21['측정일시']
air_21['time']=air_21['time'].map(lambda x : zerofrom24(x,'%Y%m%d%H'))
air_21
```

```python
air_21.dtypes
```

```python
air_22['time']=air_22['측정일시']
air_22['time']=air_22['time'].map(lambda x : zerofrom24(x,'%Y%m%d%H'))

air_22.dtypes
```

#### **<span style="color:blue">[2-6] 전일 같은 시간 변수 추가</span>**

* 모델링에 유용한 변수로 전일 같은 시간(24시간 전) 미세먼지 농도 변수를 추가합니다. 
* 시계열 데이터 처리를 위한 shift 연산을 참고하세요.

```python
# df_21, df_22의 index(time)를 month, day, hour 로 쪼개기 (year는 필요 없음)

df_21['month'] = df_21.index.month
df_21['day'] = df_21.index.day
df_21['hour'] = df_21.index.hour

df_22['month'] = df_22.index.month
df_22['day'] = df_22.index.day
df_22['hour'] = df_22.index.hour

# month, day, hour 열을 기준으로 데이터프레임 정렬
df_21 = df_21.sort_values(by=['month', 'day', 'hour'])
df_22 = df_22.sort_values(by=['month', 'day', 'hour'])
```

```python
# df_21, df_22에 전일 같은 시간 미세먼지 농도 변수(PM10_lag1) 추가
# 전일 같은 시간은 24시간 전 입니다.

df_21['PM10_lag1'] = df_21['PM10'].shift(24)

df_22['PM10_lag1'] = df_22['PM10'].shift(24)

df_21['PM10_lag1']
```

```python
# 'PM10_lag1' 열의 결측치를 12월 31일 데이터로 채우기
for i in range(1, 24):  # 1시부터 23시까지
    df_21.loc['2021-01-01 0{:02d}:00:00'.format(i), 'PM10_lag1'] = df_21.loc['2021-12-31 0{:02d}:00:00'.format(i), 'PM10']

#22년도 1월 1일 데이터는 21년도 12월 31일 데이터로 채움
for i in range(1, 24):  # 1시부터 23시까지
    df_22.loc['2022-01-01 0{:02d}:00:00'.format(i), 'PM10_lag1'] = df_21.loc['2021-12-31 0{:02d}:00:00'.format(i), 'PM10']
df_21['PM10_lag1']
```
